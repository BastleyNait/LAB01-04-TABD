```python
def realizar_consulta(texto_consulta, n_resultados=3):
    # Ejecutar la consulta en la colecciГіn
    resultados = col_superheroes.query(
        query_texts=[texto_consulta],
        n_results=n_resultados,
        include=['documents', 'distances', 'metadatas']
    )

    # Visualizar resultados con el formato solicitado
    print(f"\nрџђЌ Consulta: '{texto_consulta}'")
    print('=' * 65)

    for i, (doc, dist, meta) in enumerate(zip(
        resultados['documents'][0],
        resultados['distances'][0],
        resultados['metadatas'][0]
    ), start=1):
        similitud = 1 - dist
        print(f'\nрџђЌ Resultado #{i} | Similitud: {similitud:.4f}')
        print(f' CategorГ­a : {meta["categoria"]}')
        print(f' P. Prin   : {meta["personaje_principal"]}')
        print(f' AГ±o       : {meta["anio"]}')
        print(f' Texto     : {doc[:80]}\n {doc[80:]}')
        print('-' * 65)

# Ejemplo de uso de la funciГіn
# consulta 1
realizar_consulta('viajes en el espacio')
# consulta 2
#realizar_consulta('bromista que es un experimento fallido y busca venganza')
# consulta 3
#realizar_consulta('pelicula en la cual incluya superheroinas')
```

```python
import pandas as pd

# f) Definir los 4 pares de consultas
pares_consultas = [
    ("hГ©roes que viajan por el cosmos", "viajes intergalГЎcticos en equipo"),
    ("millonario con armadura tecnolГіgica", "Tony Stark construye traje de hierro"),
    ("caballero oscuro contra el guasГіn", "Batman enfrenta a criminal anarquista en Gotham"),
    ("mutante con garras en el futuro", "Wolverine envejecido protegiendo a una niГ±a")
]

def analizar_pares(pares):
    datos_comparativa = []

    for consulta_a, consulta_b in pares:
        # Ejecutar Consulta A
        res_a = col_superheroes.query(query_texts=[consulta_a], n_results=3)
        ids_a = set(res_a['ids'][0])
        sims_a = [1 - d for d in res_a['distances'][0]]

        # Ejecutar Consulta B
        res_b = col_superheroes.query(query_texts=[consulta_b], n_results=3)
        ids_b = set(res_b['ids'][0])
        sims_b = [1 - d for d in res_b['distances'][0]]

        # g) y i) CГЎlculos
        comunes = ids_a.intersection(ids_b)
        diferentes = ids_a.symmetric_difference(ids_b)
        jaccard = len(comunes) / len(ids_a.union(ids_b))

        datos_comparativa.append({
            'Consulta A': consulta_a,
            'Consulta B': consulta_b,
            'IDs Comunes': list(comunes),
            'IDs Diferentes': list(diferentes),
            'Simil. Prom. A': sum(sims_a) / len(sims_a),
            'Simil. Prom. B': sum(sims_b) / len(sims_b),
            'Indice Jaccard': round(jaccard, 2)
        })

    return pd.DataFrame(datos_comparativa)

# h) Construir tabla comparativa
df_resultados = analizar_pares(pares_consultas)
display(df_resultados)
```

### Conclusiones sobre la RecuperaciГіn SemГЎntica en ChromaDB

> Agregar bloque entrecomillado



El anГЎlisis de los pares de consultas revela que la recuperaciГіn vectorial de ChromaDB no depende de la coincidencia exacta de palabras, sino del contexto multidimensional. En los resultados, se observa que incluso cuando la redacciГіn varГ­a drГЎsticamente (ej. 'caballero oscuro' vs 'Batman'), el modelo de embeddings es capaz de mapear ambos conceptos a coordenadas cercanas en el espacio latente, manteniendo una coherencia en los IDs comunes. Sin embargo, la especificidad juega un rol crГ­tico: las consultas que incluyen nombres propios como 'Tony Stark' o 'Wolverine' tienden a generar puntuaciones de similitud mГЎs robustas que las descripciones genГ©ricas como 'millonario con armadura'.

El Г­ndice de Jaccard muestra que el solapamiento no siempre es perfecto (1.0), lo cual es un indicador de que matices lingГјГ­sticos sutiles pueden alterar la frontera de decisiГіn de los 'k' vecinos mГЎs cercanos. Esto demuestra que la semГЎntica permite una flexibilidad que los motores de bГєsqueda tradicionales de texto plano no poseen, permitiendo que conceptos abstractos como 'viajes intergalГЎcticos' conecten con 'cosmos' de manera natural. En conclusiГіn, para optimizar resultados en bases de datos vectoriales, es fundamental equilibrar la naturalidad del lenguaje con tГ©rminos especГ­ficos que anclen la bГєsqueda a los metadatos y documentos mГЎs relevantes, minimizando el ruido semГЎntico que puede introducir una redacciГіn demasiado vaga.

```python
# l) ImplementaciГіn de bГєsqueda avanzada
def buscar_articulos(consulta, area=None, anio_desde=None, anio_hasta=None, max_turnitin=20, top_k=5):
    filtros = []

    # Construir filtros de metadatos
    if area:
        filtros.append({"area": {"$eq": area}})

    filtros.append({"puntaje_similitud_turnitin": {"$lte": max_turnitin}})

    if anio_desde:
        filtros.append({"anio": {"$gte": anio_desde}})

    if anio_hasta:
        filtros.append({"anio": {"$lte": anio_hasta}})

    # Combinar filtros con $and si hay mГЎs de uno
    where_clause = {"$and": filtros} if len(filtros) > 1 else filtros[0]

    resultados = col_epis.query(
        query_texts=[consulta],
        n_results=top_k,
        where=where_clause
    )

    print(f"\nрџ”Ћ BГєsqueda: '{consulta}' | Filtros: Area={area}, Turnitin<={max_turnitin}%")
    print("-" * 80)

    for i, (doc, meta) in enumerate(zip(resultados['documents'][0], resultados['metadatas'][0]), 1):
        print(f"{i}. [{meta['area']}] ({meta['anio']}) {doc[:70]}... | Autor: {meta['autor']} | Turnitin: {meta['puntaje_similitud_turnitin']}%")
```

```python
# m) EjecuciГіn de pruebas con distintas combinaciones
print("Ejecutando pruebas de bГєsqueda hГ­brida...")
buscar_articulos("inteligencia artificial y aprendizaje", area="IA", max_turnitin=15)
buscar_articulos("ataques y ciberseguridad", area="Seguridad", anio_desde=2022)
buscar_articulos("redes inalambricas", max_turnitin=25, top_k=3)
buscar_articulos("blockchain y contratos", anio_hasta=2023)
```
